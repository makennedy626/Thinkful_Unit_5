# Problem 6 - Sum Square Difference

Find the difference between the sum of squares of the first one hundred natural numbers and the square of the sum.

```python
def sumsq():
    result = 0
    for i in range(1,101):
        result += i*i
    return result

def sqsum():
    total = 0
    result_sqsum = 0
    for i in range(1,101):
        total += i
    result_sqsum = total*total
    return result_sqsum

def difference():
    return sqsum() - sumsq()

print(sumsq())
print(sqsum())
print(difference())
```

# Problem 7 - 10001st prime

By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13. What is the 10,001st prime number?

```python
plist = [2]
def Primes(n, x):
    #creates a generator that contains all prime numbers from min to max
    global plist
    if 2 >= n: yield 2
    i = 3
    while i <= x:
        for p in plist:
            if i%p == 0 or p*p > i: break
        if i%p:
            plist.append(i)
            if i >= n: yield i
        i = i+2
        
def find_prime():
    i = 0
    for x in Primes(2,100000000000000000000000):
        i+=1
        if i==10001:
            return x
        
print(find_prime())
```

# Problem 9 - Special Pythagorean Triplet

There exists exactly one Pythagorean triplet for which a + b + c = 1000
Find the product abc.

```python
def triplet():
    for a in range(1,1001):
        for b in range(a+1,1001):
            c = 1000 - a - b
            if a * a + b * b == c * c:
                return(str(a * b * c))
print(triplet())
```

# Problem 10 - Summation of Primes

Find the sum of all the primes below two million.

```python
def sum_primes():
    total = 0
    for x in Primes(2,2000000):
        total += x
    return total

print(sum_primes())
```
